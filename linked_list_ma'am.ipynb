```python
class Mynode:
    def _init_(self, no):  
        self.no = no
        self.link = None

    @staticmethod
    def create_node():
        no = int(input("Enter node value: "))
        return Mynode(no)

    @staticmethod  
    def insert_node_first(s):
        new_node = Mynode.create_node()
        new_node.link = s
        return new_node

    @staticmethod
    def insert_node_last(s):
        if s is None:
            s = Mynode.create_node()
        else:
            new_node = Mynode.create_node()
            t = s
            while t.link is not None:
                t = t.link
            t.link = new_node
        return s

    @staticmethod
    def display(s):
        if s is None:
            print("No nodes")
        else:
            print("Linked list:", end=" ")
            while s is not None:
                print(s.no, end=" ")
                s = s.link
            print()
        return s

    @staticmethod
    def delete_first_node(s):
        if s is None:
            print("No nodes")
        else:
            s = s.link
            print("Deleted first node") 
        return s

    @staticmethod   
    def delete_last_node(s):
        if s is None:
            print("No nodes")
        elif s.link is None:
            s = None
            print("Deleted last node")
        else:
            t = s
            while t.link.link is not None:
                t = t.link
            t.link = None
            print("Deleted last node")
        return s

    @staticmethod
    def insert_between_nodes(s):
        if s is None or s.link is None:
            print("List too short to insert between nodes.")
            return s
        pos = int(input("Enter position to insert after (1-based index): "))
        new_node = Mynode.create_node()
        temp = s
        i = 1
        while i < pos and temp is not None:
            temp = temp.link
            i += 1
        if temp is None:
            print("Position out of range.")
        else:
            new_node.link = temp.link
            temp.link = new_node
        return s

    @staticmethod
    def count_node(s):
        count = 0
        while s is not None:
            count += 1
            s = s.link
        print("Total nodes:", count)
        return count

    @staticmethod
    def reverse_node(s):
        prev = None
        current = s
        while current is not None:
            next_node = current.link
            current.link = prev
            prev = current
            current = next_node
        print("Linked list reversed.")
        return prev

# Main part of the code
s = None  
while True:
    print("\n1. Insert node at first position")
    print("2. Insert node at last position")
    print("3. Insert node between 2 nodes")
    print("4. Display nodes")
    print("5. Delete first node")
    print("6. Delete last node")
    print("7. Count nodes")
    print("8. Reverse linked list")
    print("9. Exit")

    ch = int(input("Enter your choice: "))
    if ch != 0:
        print("Option chosen:", ch)
        if ch == 1:
            s = Mynode.insert_node_first(s)
        elif ch == 2:
            s = Mynode.insert_node_last(s)
        elif ch == 3:
            s = Mynode.insert_between_nodes(s)
        elif ch == 4:
            Mynode.display(s)
        elif ch == 5:
            s = Mynode.delete_first_node(s)
        elif ch == 6:
            s = Mynode.delete_last_node(s)
        elif ch == 7:
            Mynode.count_node(s)
        elif ch == 8:
            s = Mynode.reverse_node(s)
        elif ch == 9:
            print("Exiting program.")
            break
        else:
            print("Invalid choice")
    else:
        print("Invalid input")
```
