```python
class Node:
    def __init__(self,no):
        self.no=no
        self.link=None
    @staticmethod
    def create_node():
        no=int(input("Enter node value: "))
        temp=Node(no)
        return temp
    #inserting a node in the first position
    @staticmethod  
    def insert_node_first(s):
        new_node=Node.create_node()
        new_node.link=s
        return new_node
    @staticmethod
    def insert_node_last(s):
        if s==None:
            s=Node.create_node()
        else:
            new_node=Node.create_node()
            t=s
            while(t.link!=None):
                t=t.link
            t.link=new_node
    @staticmethod
    def display(s):
        if s==None:
            print("No nodes")
        while s!=None:
            print(s.no,end=" ")
            s=s.link
    @staticmethod
    def delete_first_node(s):
        if s==None:
            print("No nodes")
        else: 
            s=s.link
            print("Deleted first node") 
    @staticmethod   
    def delete_last_node(s):
        if s==None:
            print("No nodes")
        else:
            t=s
            while(t.link.link!=None):
                t=t.link
            t.link=None
            print("Deleted last node")  
    @staticmethod
    def insert_between_nodes(s):
        if s==None:
            print("No nodes")
        else:
            no=int(input("Enter node value: "))
            new_node=Node(no)
            t=s
            while(t.link!=None):
                t=t.link
            t.link=new_node
            print("Inserted node between nodes")
            
        
    
#main part of the code where we take the input
s=None  
while True:
    print("\n1. Insert node at first position")
    print("2. Insert node at last position ")
    print("3. Insert nodes between 2 other nodes")
    print("4. Display nodes")
    print("5. Exit")
    print()
    ch=int(input("Enter your choice: "))
    if ch!=0:
        print("option chosen:",ch)
        if ch==1:
            s=Node.insert_node_first(s)
        elif ch==2:
            Node.insert_node_last(s)
        elif ch==3:
            Node.insert_between_nodes(s)
        elif ch==4:
            s=Node.display(s)
        elif ch==5:
            break
        else:
            print("Invalid choice")
    else:
        print("Invalid")
```
