Stack implementation

```python
#without size
class stack:
    def __init__(self):
        self.s=[]
    def push(self, n):
        self.s.append(n)
    def pop(self):
        self.s.pop()
    def peek(self):
        print(self.s[0])
    def isEmpty(self):
        if len(self.s)==0:
            return True
        else:
            return False
    def show(self):
        print(self.s)
    def delete(self):
        self.s.clear()
    
    
        
o=stack() #stack is created
while True:
    print("1.Push\n2.Pop\n3.Peek\n4.isEmpty\n5.Show\n6.Delete\n7.Exit")
    print(" ")
    ch=int(input("Enter your choice:"))
    print("Choice:",ch)
    if ch==1:
        n=int(input("Enter a variable to push:"))
        o.push(n)
    elif ch==2:
        o.pop()
    elif ch==3:
        o.peek()
    elif ch==4:
        print(o.isEmpty())
    elif ch==5:
        o.show()
    elif ch==6:
        o.delete()
    elif ch==7:
        print("Exiting")
        break
    else:
        print("Invalid")
        
        
    
    
    
    
    
    
```

```python
#with size
class stack:
    def __init__(self,size):
        self.s=[]
        self.size=size
    def isFull(self):
        return len(self.s)==size
    def push(self, n):
        if self.isFull():
            print("Stack full")
        else:
            self.s.append(n)
    def pop(self):
        self.s.pop()
    def peek(self):
        print(self.s[0])
    def isEmpty(self):
        if len(self.s)==0:
            return True
        else:
            return False
    def show(self):
        print(self.s)
    def delete(self):
        self.s.clear()    
    
    
size=int(input("Size:"))  
o=stack(size) #stack is created
while True:
    print("1.Push\n2.Pop\n3.Peek\n4.isEmpty\n5.Show\n6.Delete\n7.Exit")
    print(" ")
    ch=int(input("Enter your choice:"))
    print("Choice:",ch)
    if ch==1:
        n=int(input("Enter a variable to push:"))
        o.push(n)
    elif ch==2:
        o.pop()
    elif ch==3:
        o.peek()
    elif ch==4:
        print(o.isEmpty())
    elif ch==5:
        o.show()
    elif ch==6:
        o.delete()
    elif ch==7:
        print("Exiting")
        break
    else:
        print("Invalid")
        
        
    
    
    
    
    
    

```

Tower of hanoi

```python
class Tower:
    def __init__(self):
        print("TOWER OF HANOI")
        print()
        print("Given problem    A=[3,2,1]    B=[]    C=[]")
        print("Expected Output  A=[]         B=[]    C=[1,2,3]")
        self.A=[3,2,1]
        self.B=[]
        self.C=[]        
    def working(self):
        while len(self.A)==3:
            x=self.A.pop()
            if self.A[-1]>self.B[-1]:
                self.B.append(x)
            
            print(self.A,"    ",self.B,"    ",self.C)
            
            
            
o=Tower()
o.working()

```
